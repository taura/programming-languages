* Write a function `gaussian_density` (or `GaussianDensity`, according to your language's case convention) that takes `x`, `mu` (mean), and `sigma` (standard deviation, you may assume `sigma > 0`) and returns the value of the Gaussian (normal) probability density function at `x`:

$$
\frac{1}{ \sqrt{2\pi}\,\sigma } \exp\!\left(-\frac{ (x-\mu)^2} {2\sigma^2} \right)
$$
