
# 3. Your Answer (assembly)
* Running the cell below writes the skeleton assembly file `ax_by_cz_long.s`.
* Fill in your instructions after the line `// ------- write your answer here -------`, then run the cell again to save it.

````python
%%writefile_ ax_by_cz_long.s
	.arch armv8-a
	.file	"ax_by_cz_long.c"
	.text
	.align	2
	.p2align 4,,11
	.global	ax_by_cz_long
	.type	ax_by_cz_long, %function
ax_by_cz_long:
.LFB0:
	.cfi_startproc
	// ------- write your answer here -------
	.cfi_endproc
.LFE0:
	.size	ax_by_cz_long, .-ax_by_cz_long
	.ident	"GCC: (Ubuntu 13.3.0-6ubuntu2~24.04) 13.3.0"
	.section	.note.GNU-stack,"",@progbits
````


# 4. Checker
* The following C program calls your `ax_by_cz_long` function and checks the result against a reference computed in C.

````python
%%writefile_ check_ax_by_cz_long.c
#include <assert.h>
#include <stdio.h>
#include <stdlib.h>
#include <math.h>
long ax_by_cz_long(long a, long x, long b, long y, long c, long z);

int main(int argc, char ** argv) {
  assert(argc == 7);
  long a = atol(argv[1]);
  long x = atol(argv[2]);
  long b = atol(argv[3]);
  long y = atol(argv[4]);
  long c = atol(argv[5]);
  long z = atol(argv[6]);
  long r = ax_by_cz_long(a, x, b, y, c, z);
  long rc = a * x + b * y + c * z;
  if (r == rc) {
    printf("OK %ld %ld\n", r, rc);
    return 0;
  } else {
    printf("NG %ld %ld\n", r, rc);
    return 1;
  }
}
````
