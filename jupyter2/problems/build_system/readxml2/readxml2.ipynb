
# <font color="green">Building multiple files</font>

* Large programs rarely fit in a single file
* Splitting a program into multiple files requires understanding how to refer to names across files
* Learn what it takes to reference names defined in other files

# Go

## Basics

* In Go, splitting a program across multiple files in the same directory is straightforward
* A file can reference names in another file in the same directory as if they were in the same file
* All that is required is that every file has the same `package `_package-name_ clause at the top (for an executable program, this is `package main`)
* Things get more complex when files are in separate directories, a topic we will not cover further here
* For example, to put function `readxml` in `readxml.go` and call it from `readxml2.go`, all you need is `package main` at the top of both files

## General concepts behind it

* In Go, a directory = a _package_
* You can refer to any name in the same package (i.e., the same directory), even if it is in another file, exactly as if it were in the same file
* To refer to a name in another package, you must import the package:
```
import "module-name/package-path"
```
* See [Tutorial: Create a Go module](https://go.dev/doc/tutorial/create-module) for details

# Julia

## Basics

* You can split a program into multiple files; a file can load another file with `include(`_file_`)`, whose effect is as if the contents of _file_ were pasted in place
* For example, to put function `readxml` in `readxml.jl` and call it from `readxml2.jl`, put `include("readxml.jl")` in `readxml2.jl`
* You can then call `readxml` as if it were defined in the same file

## More general concepts behind it

* The effect of `include` is verbatim textual insertion of the specified file
* To avoid name conflicts, you can put definitions into separate _modules_
* A module is a mechanism for grouping names into separate namespaces, and is largely orthogonal to file structure: you can define multiple modules in a single file, and split a single module across multiple files using `include`
* The syntax for defining a module is:
```
module ModuleName
  ... definitions ...
end
```
* Definitions inside a `module ... end` block belong to that module
* Only names explicitly listed in an `export` statement within the module are _exported_
* To refer to a name from outside the module, qualify it with the module name:
```
module Foo
f(x) = x + 1
end
Foo.f(3)
```
* The effect of `import `_module_ is:
  * If the module is not already defined, Julia looks for a file _module_`.jl` in the directories listed in `LOAD_PATH`
  * All top-level definitions in that file are treated as belonging to _module_
  * All names in _module_ become accessible as _module_._name_
* The effect of `using `_module_ is that of `import `_module_, plus bringing all _exported_ names of _module_ into scope directly, without the module qualifier

# OCaml

## Basics

* To refer to a name defined in another `.ml` file in the _same directory_, prefix the name with the capitalized filename (without extension)
* For example, to put function `readxml` in `readxml.ml` and call it from `main.ml` in the same directory, reference it as `Readxml.readxml`

## More general concepts behind it

* OCaml uses _modules_ to separate names into namespaces
* Each `.ml` file automatically defines a module (`foo.ml` defines a module `Foo`), and all definitions in the file belong to that module
* The mechanism by which the OCaml compiler locates other modules when compiling is complex; we leave it to `dune` to handle

# Rust

## Basics

Say `main.rs` wants to use a function `parse` in `foo.rs`:
* Put
```
mod foo;
```
in `main.rs`
* Define the function and mark it _public_ in `foo.rs` by putting the `pub` keyword before it:
```
pub fn parse(...) { ... }
```
* Then either:
  * Call `foo::parse` in `main.rs`, or
  * Put `use foo::parse;` near the top of `main.rs` and call `parse` directly

## Concepts behind this

* In Rust, a package (created by `cargo new`) initially contains a single _crate_ (you may add more crates to a package, but we will not cover that here)
* A crate consists of multiple _modules_ forming a hierarchical tree
* A module in crate $C$ is uniquely identified by its path from the _crate root_, written as $C$::$M_1$::$M_2$::...::$M_n$, where $M_n$ is a child of $M_{n-1}$, ..., which is a child of $M_1$, which is a child of the crate root
* A name defined in a module becomes _public_ (visible from outside the module) when it is prefixed with the `pub` keyword:
```
pub fn parse(...) { ... }
```
* The top-level definitions in `src/main.rs` (generated by `cargo new`) belong to the crate root
* There are two ways to define a child module:
  1. An _inline_ module definition inside the parent:
```
mod module_name {
  ...
}
```
  2. A declaration in the parent:
```
mod module_name;
```
with the module's definitions in a separate file at the appropriate path. Specifically, module $C$::$M_1$::$M_2$::...::$M_n$ should be placed in `src/`$M_1$`/`$M_2$`/.../`$M_n$`.rs` (new convention, supported since Rust 2018) or `src/`$M_1$`/`$M_2$`/.../`$M_n$`/mod.rs` (old convention). Both styles are valid and can be mixed within a crate. Note that this path convention assumes every ancestor module also uses the external file convention; if an ancestor is defined inline, the file is located relative to the file containing that inline definition.
* Within a crate, you can refer to a public name in another module by qualifying it with the module path, either as an _absolute_ path or a _relative_ path:
  * `crate` is a keyword referring to the crate root (used in absolute paths)
  * `super` is a keyword referring to the parent module (used in relative paths)
* Similarly <font color="blue">use $C$::$M_1$::$M_2$:: ... ::$M_n$::_name_</font>, where _name_ is a name defined in module $C$::$M_1$::$M_2$:: ... ::$M_n$, brings _name_ into scope so it can be used without the full qualifier; _name_ may itself be a module name
* In the special case where _name_ is `*`, <font color="blue">use $C$::$M_1$::$M_2$:: ... ::$M_n$::*</font> brings all public names in that module into scope directly (glob import)
