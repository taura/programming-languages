# 2. Go


## 2-1. AI Tutor

````python
import heytutor
````


## 2-2. Source Code

````python
%%writefile_ go/funcall.go
package how_compiled
import "math"

func CallTanh(x float64) float64 {
	return math.Tanh(x + 1.0) + x
}
````


## 2-3. Generate Assembly Code

````python
%%bash_
export PATH=${PATH}:~/.local/go/bin:~/go/bin
go build -o go/funcall.o go/funcall.go
go tool objdump -gnu go/funcall.o > go/funcall.s
cat go/funcall.s
````


## 2-4. Visualize Assembly Code (Control Flow Graph)
* You can visualize the control flow graph of the assembly code.


````python
import vis_asm
vis_asm.vis_asm_file("go/funcall.s")
````
