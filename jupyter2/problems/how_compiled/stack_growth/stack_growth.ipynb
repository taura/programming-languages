# <font color="green">How the Stack Grows</font>

- Investigate what happens to the stack during deep recursive calls.
- Define a recursive function `sum_array_rec` (or `SumArrayRec` in Go) that takes an array $a$ and an integer $n$ and computes:

$$\text{sum}(a, n) = \begin{cases} 0 & (n = 0) \\ \text{sum}(a,\, n-1) + a[n-1] & (n > 0) \end{cases}$$

In Julia, use 1-based indexing:

$$\text{sum}(a, n) = \begin{cases} 0 & (n = 0) \\ \text{sum}(a,\, n-1) + a[n] & (n > 0) \end{cases}$$

## Problems

Make a note of the following:

1. How much stack space is allocated per function call, and how is it allocated?
2. What values are saved there?

# 2. Go


## 2-1. AI Tutor

````python
import heytutor
````


## 2-2. Source Code

````python
%%writefile_ go/stack_growth.go
package how_compiled

func SumArrayRec(a []float64, n int64) float64 {
	if (n == 0) {
		return 0.0
	} else {
		return SumArrayRec(a, n - 1) + a[n]
	}
}
````


## 2-3. Generate Assembly Code

````python
%%bash_
export PATH=${PATH}:~/.local/go/bin:~/go/bin
go build -o go/stack_growth.o go/stack_growth.go
go tool objdump -gnu go/stack_growth.o > go/stack_growth.s
cat go/stack_growth.s
````

````python
%%writefile_ jl/stack_growth.jl
function sum_array_loop(a :: Vector{Float64}, n :: Int64) :: Float64
    s = 0.0
    for i in 1:n
        s += a[i]
    end
    return s
end

import InteractiveUtils
InteractiveUtils.code_native(sum_array_loop)
````

````python
%%writefile_ ml/stack_growth.ml
let rec sum_array_rec a n =
  if n = 0 then
    0.0
  else
    (sum_array_rec a (n - 1)) +. a.(n - 1)
;;
````
