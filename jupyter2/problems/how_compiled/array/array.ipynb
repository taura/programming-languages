# 2. Go


## 2-1. AI Tutor

````python
import heytutor
````


## 2-2. Source Code

````python
%%writefile_ go/array.go
package how_compiled

func GetArrayElem(a []float64, i int64) float64 {
    return a[i]
}
````


## 2-3. Generate Assembly Code

````python
%%bash_
export PATH=${PATH}:~/.local/go/bin:~/go/bin
go build -o go/array.o go/array.go
go tool objdump -gnu go/array.o > go/array.s
cat go/array.s
````


## 2-4. Visualize Assembly Code (Control Flow Graph)

````python
import vis_asm
vis_asm.vis_asm_file("go/array.s")
````

````python
%%writefile_ jl/array.jl
function get_array_elem(a :: Vector{Float64}, i :: Int64) :: Float64
    a[i]
end

import InteractiveUtils
InteractiveUtils.code_native(get_array_elem)
````
