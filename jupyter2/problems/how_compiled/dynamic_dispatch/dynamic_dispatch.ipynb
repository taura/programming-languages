# 2. Go


## 2-1. AI Tutor

````python
import heytutor
````


## 2-2. Source Code

````python
%%writefile_ go/dynamic_dispatch.go
package how_compiled

type Shape interface {
	area() float64
}

func CallArea(s Shape) float64 {
	return s.area()
}
````


## 2-3. Generate Assembly Code

````python
%%bash_
export PATH=${PATH}:~/.local/go/bin:~/go/bin
go build -o go/dynamic_dispatch.o go/dynamic_dispatch.go
go tool objdump -gnu go/dynamic_dispatch.o > go/dynamic_dispatch.s
cat go/dynamic_dispatch.s
````


# 5. Rust


## 5-1. AI Tutor

````python
import heytutor
````


## 5-2. Source Code

````python
%%writefile_ rs/dynamic_dispatch.rs
trait Shape {
    fn area(&self) -> f64;
}

#[no_mangle]
fn call_area(s : &dyn Shape) -> f64 {
    s.area()
}
````


## 5-3. Generate Assembly Code

````python
%%bash_
. ~/.cargo/env
rustc -C opt-level=3 --emit asm --crate-type lib rs/dynamic_dispatch.rs -o rs/dynamic_dispatch.s
cat -n rs/dynamic_dispatch.s
````
