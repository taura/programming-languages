
## 2-2. Source Code

````python
%%writefile_ go/struct.go
package how_compiled

type Point struct {
    x float64
    y float64
}

func GetPointY(p Point) float64 {
    return p.y
}
````


## 2-3. Generate Assembly Code

````python
%%bash_
export PATH=${PATH}:~/.local/go/bin:~/go/bin
go build -o go/struct.o go/struct.go
go tool objdump -gnu go/struct.o > go/struct.s
cat go/struct.s
````


# 5. Rust


## 5-1. AI Tutor

````python
import heytutor
````


## 5-2. Source Code

````python
%%writefile_ rs/struct.rs
#![allow(dead_code)]

struct Point {
    x : f64,
    y : f64
}

#[no_mangle]
fn get_point_y(p : Point) -> f64 {
    p.y
}
````


## 5-3. Generate Assembly Code

````python
%%bash_
. ~/.cargo/env
rustc -C opt-level=3 --emit asm --crate-type lib rs/struct.rs -o rs/struct.s
cat -n rs/struct.s
````
