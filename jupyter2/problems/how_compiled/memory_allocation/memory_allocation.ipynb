# 2. Go


## 2-1. AI Tutor

````python
import heytutor
````


## 2-2. Source Code

````python
%%writefile_ go/memory_allocation.go
package how_compiled

type Point struct {
    x float64
    y float64
}

func mkPoint(x, y float64) Point {
	p := Point{x: x + 1.0, y: y + 2.0}
	return p
}
````


## 2-3. Generate Assembly Code

````python
%%bash_
export PATH=${PATH}:~/.local/go/bin:~/go/bin
go build -o go/memory_allocation.o go/memory_allocation.go
go tool objdump -gnu go/memory_allocation.o > go/memory_allocation.s
cat go/memory_allocation.s
````


## 2-4. Visualize Assembly Code (Control Flow Graph)
* You can visualize the control flow graph of the assembly code.


````python
import vis_asm
vis_asm.vis_asm_file("go/memory_allocation.s")
````
