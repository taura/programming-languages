# 2. Go


## 2-1. AI Tutor

````python
import heytutor
````


## 2-2. Source Code

````python
%%writefile_ go/loop.go
package how_compiled

func SumArrayLoop(a []float64, n int64) float64 {
	s := 0.0
	var i int64;
	for i = 0; i < n; i++ {
		s += a[i]
	}
	return s;
}
````


## 2-3. Generate Assembly Code

````python
%%bash_
export PATH=${PATH}:~/.local/go/bin:~/go/bin
go build -o go/loop.o go/loop.go
go tool objdump -gnu go/loop.o > go/loop.s
cat go/loop.s
````


## 2-4. Visualize Assembly Code (Control Flow Graph)

````python
import vis_asm
vis_asm.vis_asm_file("go/loop.s")
````


# 4. OCaml


## 4-1. AI Tutor

````python
import heytutor
````


## 4-2. Source Code

````python
%%writefile_ ml/loop.ml
let sum_array_loop a n =
  let s = ref 0.0 in
  for i = 0 to n do
    s := !s +. a.(i)
  done;
  !s
;;
````


## 4-3. Generate Assembly Code

````python
%%bash_
eval $(opam env)
ocamlopt -O3 -S ml/loop.ml
cat -n ml/loop.s
````


## 5-2. Source Code

````python
%%writefile_ rs/loop.rs
#[no_mangle]
fn sum_array_loop(a : &[f64], n : usize) -> f64 {
    let mut s = 0.0;
    for i in 0..n {
	s += a[i];
    }
    s
}
````


## 5-3. Generate Assembly Code

````python
%%bash_
. ~/.cargo/env
rustc -C opt-level=3 --emit asm --crate-type lib rs/loop.rs -o rs/loop.s
cat -n rs/loop.s
````
