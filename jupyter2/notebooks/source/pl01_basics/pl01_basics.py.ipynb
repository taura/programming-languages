# 6. How to Use AI tutor

## 6-1. `%%hey`
* By executing a cell beginning with `%%hey` you can ask a question to AI

````python
%%hey

How to write a function in C++?
````


* `%%hey` can be followed by `problem_file=...` and `answer_file=...`
* `{{problem}}` in the cell is automatically replaced by the contents of the file specified with `problem_file=`
* `{{answer}}` in the cell is automatically replaced by the contents of the file specified with `answer_file=`
* For example,
* A hint on this specific problem
```
%%hey problem_file=hello.md
I am working on Rust. Give me a hint on this problem.
{{problem}}
```
* In each problem notebook, `{{problem}}` is actually in the system prompt, so you can omit it.  But if you are unsure, it is safe to make it explicit
```
%%hey problem_file=hello.md
I am working on Rust. Give me a hint on this problem.
```
* Help when you struggle
```
%%hey problem_file=hello.md answer_file=rs/hello.rs
I get this error when I compile it. What's wrong?"

My program:
{{answer}}

Error message:
  <paste the compile error message> 

```
* You are encouraged to ask a feedback once you think you are done with the problem, to know if there is a better answer.  You can do so by something like:
```
%%hey problem_file=hello.md answer_file=rs/hello.rs
My Answer:
{{answer}}

Give me a feedback to my answer.
```

## 6-2. `%%bash_` and `%%writefile_`
* There are `%%bash_` and `%%writefile_` that are similar to `%%bash` and `%%writefile`, respectively, but extended to work with AI tutor
* Problem notebooks have many such cells to help you execute commands quickly

````python
%%writefile_ hello.c
/* a Python cell that just saves the contents into a file (hello.c) */
#include <stdio.h>
int main() {
    printf("hello\n");
    return 0;
}
````

````python
%%bash_
gcc hello.c -o hello
./hello
````
