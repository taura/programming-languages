# 2. How to let the C compiler generate assembly code
* `gcc`/`g++` is a compiler for C/C++.
* The `-S` option emits assembly code and stops after that.
* For example, to get the assembly code for `abc.c`, 
```
gcc -O3 -S abc.c
cat abc.s
```
